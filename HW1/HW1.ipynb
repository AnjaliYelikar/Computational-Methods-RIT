```python
#PROBLEM 1

#BISECTION METHOD

import matplotlib.pyplot as plt

# Defining the function/expression here

f = lambda x: (x**2) - 45. 
#f(2)

def BisectionMethod(f,a,b,convlimit):
    # This function finds the root of a function 'f' within the given limits 'a' and 'b' 
    # till it reaches the set convergence criterion 'convlimit' by the Bisection method
    
    
    error=1.
    i=0.
    while (error >= convlimit):
        r = (a+b)/2.
        if (f(a)*f(r)) <= 0.:
            b=r
            f(b)==f(r)
        else:
            a=r
            f(a)==f(r)
        error=abs((b-a))
        i=i+1
        print(i,r,f(r),error)
        plt.scatter(i,r)
        plt.xlabel('iteration')
        plt.ylabel('root')
        plt.savefig('Bisection.pdf')
    print("The root is=", r)
    
    
BisectionMethod(f,6.,7.,10.**-7)    
print
```

```python
#NEWTON METHOD

import matplotlib.pyplot as plt
from sympy import *
x=symbols("x")

f = lambda x: (x**2) - 45.          
df = lambdify(x,diff(f(x),x))
#d2f = lambdify(x,diff(df(x),x))
#print(df(2.))

def NewtonMethod(f,df,a,convlimit):
    # This function finds the root of a function 'f' within the given limits 'a' and 'b' 
    # till it reaches the set convergence criterion 'convlimit' by the Newton method
    error=1.
    i=0.
    while(error >= convlimit):
        a = a- (f(a)/df(a))
        error=abs((f(a)/df(a))/a)
        #error=d2f(a)/2.
        i=i+1
        print(i,a,f(a),error)
        plt.scatter(i,a)
        plt.xlabel('iteration')
        plt.ylabel('root')
        plt.savefig('Newton.pdf')
    print("The root is =", a)
        
NewtonMethod(f,df,7.,10.**-7)
```
